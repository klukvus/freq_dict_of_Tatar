## Сравнение частотных списков

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import itertools
```

```python
# Загрузка данных
def load_and_prepare(filepath):
    df = pd.read_csv(filepath, decimal=',')
    df['lemma_POS'] = df['lemma'] + '_' + df['POS']
    df['ipm'] = pd.to_numeric(df['ipm'].replace(',', '.'), errors='coerce').fillna(0)
    df = df[['lemma_POS', 'ipm']]
    return df
```

```python
# Загрузка всех списков
filepaths = {
    'corpus1': 'freq_PKT.csv',
    'corpus2': 'freq_SE.csv',
    'corpus3': 'Leipzig.csv',
}

dataframes = {name: load_and_prepare(path) for name, path in filepaths.items()}
```

```python
# Объединение всех списков
combined_df = pd.DataFrame()
for name, df in dataframes.items():
    df = df.rename(columns={'ipm': f'ipm_{name}'})
    if combined_df.empty:
        combined_df = df
    else:
        combined_df = pd.merge(combined_df, df, on='lemma_POS', how='outer')

# Заполнение отсутствующих значений нулями
combined_df.fillna(0, inplace=True)
```

```python
combined_df
```

```python
# Расчет log-likelihood для каждой пары корпусов
def compute_llr(freqs1, freqs2):
    llr_values = []
    for f1, f2 in zip(freqs1, freqs2):
        if f1 == 0 and f2 == 0:
            llr = 0
        else:
            total = f1 + f2
            E1 = total * (sum(freqs1) / (sum(freqs1) + sum(freqs2)))
            E2 = total * (sum(freqs2) / (sum(freqs1) + sum(freqs2)))
            llr = 2 * ((f1 * np.log(f1 / E1) if f1 != 0 else 0) +
                       (f2 * np.log(f2 / E2) if f2 != 0 else 0))
        llr_values.append(llr)
    return llr_values
```

```python
# Применение log-likelihood ко всем парам
corpus_names = list(filepaths.keys())
results = []

for corpus_a, corpus_b in itertools.combinations(corpus_names, 2):
    freqs_a = combined_df[f'ipm_{corpus_a}']
    freqs_b = combined_df[f'ipm_{corpus_b}']
    llr = compute_llr(freqs_a, freqs_b)
    temp_df = pd.DataFrame({
        'lemma_POS': combined_df['lemma_POS'],
        f'LLR_{corpus_a}_vs_{corpus_b}': llr
    })
    results.append(temp_df)
```

```python
# Объединение всех результатов
llr_df = results[0]
for df in results[1:]:
    llr_df = pd.merge(llr_df, df, on='lemma_POS')
```

```python
# Анализ результатов
# Определение корпуса с максимальной частотой для каждой леммы
def determine_dominant_corpus(row):
    freqs = {name: row[f'ipm_{name}'] for name in corpus_names}
    dominant = max(freqs, key=freqs.get)
    return dominant

combined_df['dominant_corpus'] = combined_df.apply(determine_dominant_corpus, axis=1)

```

```python
# Объединение с LLR результатами
final_df = pd.merge(combined_df, llr_df, on='lemma_POS')

# Сортировка по значимости различий (можно выбрать нужный столбец LLR)
final_df.sort_values(by='LLR_corpus1_vs_corpus2', ascending=False, inplace=True)

# Сохранение результатов
final_df.to_csv('comparison_results.csv', index=False)
```
